```python
import pandas as pd
import numpy as np

edu2 = pd.read_csv('C:/Users/olivi/Documents/Ironhack/Project Primary Education/1_Data/edu2.csv')
edu2
```

```python
population = pd.read_csv('C:/Users/olivi/Documents/Ironhack/Project Primary Education/1_Data/countries_by_population_2019.csv')
ISO3 = pd.read_csv('C:/Users/olivi/Documents/Ironhack/Project Primary Education/1_Data/country_codes_2020.csv')

popISO3 = pd.merge(population, ISO3, how='left', on="name")

popISO3
```

```python
popISO3.rename(columns={'cca3': 'ISO3'}, inplace=True)
popISO3
```

```python
edu4 = pd.merge(popISO3, edu2, how='right', on="ISO3")

edu4['pop2019']=edu4['pop2019']
edu4['pop2019']=edu4['pop2019']
edu4
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from numpy.polynomial.polynomial import polyfit

# Set the figure size
plt.figure(figsize=(10, 10))

# Scatterplot
plt.scatter(
    x = edu4['Residence Urban'], 
    y = edu4['Residence Rural'], 
    s= edu4['pop2019']/100, 
    c= edu4['Region'].cat.codes, 
    cmap="Accent", 
    alpha=0.6, 
    edgecolors="white", 
    linewidth=2)


# Add titles (main and on axis)
plt.xlabel("Access to Primary Education for Urban dwellers")
plt.ylabel("Access to Primary Education for Rural dwellers")
plt.title("Access to Primary Education worldwide")
plt.ylim(0,1)
plt.xlim(0,1)

ax = plt.gca()
ax.annotate('IND', (0.14, 0.04)) 
ax.annotate('PAK', (0.18, 0.04)) 
ax.annotate('RUS',(0.9,0.82) )
ax.annotate('JPN',(0.74,0.81) )
ax.annotate('IRQ',(0.53,0.33) )
ax.annotate('IDN',(0.24,0.09) )
ax.annotate('BGD',(0.49,0.3) )
ax.annotate('THA',(0.75,0.65) )
ax.annotate('CHN', (0.91,0.53))


# Fit with polyfit
x = [0.6, 0,8]
plt.plot(x, x, '-')
plt.show()
```
